```python

import numpy as np
def random_dice(N):
    x = np.random.rand(N)*6 + 1 
    y = np.floor(x).astype(int)
    return y
print(random_dice(20))

def reshape_vector(y):
    x =  y.reshape(-1,2)
    return x
y = np.array([1, 2, 3, 4, 5, 6])
print(reshape_vector(y))

def max_value(z):
    x = np.max(z)
    y = np.where(z == x)
    return y

z = np.array([[1,2],[3,4],[5,6]])
print(max_value(z))

def count_ones(v):
    x = v.count(1)
    return x

v = [1,2,3,4,5,6,1,2,1,1]
print(count_ones(v))
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
from matplotlib.colors import LogNorm 
      
dx, dy = 0.015, 0.05
x = np.arange(-4.0, 4.0, dx) 
y = np.arange(-4.0, 4.0, dy) 
X, Y = np.meshgrid(x, y) 
   
extent = np.min(x), np.max(x), np.min(y), np.max(y) 
   
Z1 = np.add.outer(range(8), range(8)) % 2
plt.imshow(Z1, cmap ="binary_r", interpolation ='nearest', 
                               extent = extent, alpha = 1) 
   
def geeks(x, y): 
    return (1 - x / 2 + x**5 + y**6) * np.exp(-(x**2 + y**2)) 
   
Z2 = geeks(X, Y) 
   
plt.imshow(Z2, cmap ="Blues", alpha = 0.7,  
           interpolation ='bilinear', extent = extent) 
  
plt.title('matplotlib.pyplot.imshow() function Example',  
                                     fontweight ="bold") 
plt.show() 
```
